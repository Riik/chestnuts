```python
import folium
```

```python
import json
with open('Bomen_in_beheer_door_gemeente_Delft.geojson', 'r') as f:
    raw = f.read()
data = json.loads(raw)
data['features'][0]

props = [dict(item['properties'], **item['geometry']) for item in data['features']]

props[0]
```

```python
import pandas as pd
df = pd.DataFrame(props)
def make_info_string(row):
    return f"""{row['BOOMSOORT1']} ({row['BOOMSOORT_']})
Plantjaar: {row['PLANTJAAR']}
Hoogte(m): {row['BOOMHOOGTE']}"""
df['lat'] = df['coordinates'].map(lambda x: x[0])
df['lon'] = df['coordinates'].map(lambda x: x[1])
df['info'] = df.apply(make_info_string, axis=1)
df
```

```python
df_chestnut = df[df['BOOMSOORT1'].str.contains("kastanje") & ~df['BOOMSOORT_'].str.contains("Castanea")]
df_sweet_chestnut = df[df['BOOMSOORT_'].str.contains("Castanea")]
df_hazelaar = df.copy()[df['BOOMSOORT_'].str.contains("avellana")]
trees = [
    ("Alle kastanjes", df_chestnut, 'green'),
    ("Tamme kastanjes", df_sweet_chestnut, 'blue'),
    ("Hazelaar", df_hazelaar, 'red')
]
```

```python
from folium.plugins import MarkerCluster
m = folium.Map(location=[51.9940992,4.3522975], max_zoom=19, zoom_start=13, prefer_canvas=True)

def new_marker_cluster(df, name, color):
    marker_cluster = MarkerCluster(maxClusterRadius=40, name=name)
    for idx, row in df.iterrows():
        folium.CircleMarker(
            [row['lon'], row['lat']], 
            popup=row['info'],
            color='grey',
            fill=True,
            fill_color=color,
        ).add_to(marker_cluster)
    return marker_cluster

for (name, df, color) in trees:
    new_marker_cluster(df, name, color).add_to(m)
    
folium.LayerControl().add_to(m)

m
```

```python
m.save('index.html')
```

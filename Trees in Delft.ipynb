```python
import folium
```

```python
import json
with open('Bomen_in_beheer_door_gemeente_Delft.geojson', 'r') as f:
    raw = f.read()
data = json.loads(raw)
data['features'][0]

props = [dict(item['properties'], **item['geometry']) for item in data['features']]

props
```

```python
import pandas as pd
df = pd.DataFrame(props)
def make_info_string(row):
    return f"""{row['BOOMSOORT1']} ({row['BOOMSOORT_']})
Plantjaar: {row['PLANTJAAR']}
Hoogte(m): {row['BOOMHOOGTE']}"""
df['lat'] = df['coordinates'].map(lambda x: x[0])
df['lon'] = df['coordinates'].map(lambda x: x[1])
df['info'] = df.apply(make_info_string, axis=1)
df = df[df['BOOMSOORT1'].str.contains("kastanje")]
df
```

```python
from folium.plugins import MarkerCluster
m = folium.Map(location=[51.9940992,4.3522975], max_zoom=19, prefer_canvas=True)
marker_cluster = MarkerCluster(maxClusterRadius=40).add_to(m)

for idx, row in df.iterrows():
    folium.CircleMarker([row['lon'], row['lat']], popup=row['info']).add_to(marker_cluster)
    
m
```

```python
m.save('index.html')
```
